# EDA & Exploration

This notebook performs quick exploratory data analysis on the processed credit dataset.
It loads `data/processed/train.parquet` and `data/processed/test.parquet`, shows basic summaries,
and produces visual checks for class balance and missingness.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
%matplotlib inline

train_path = Path('data/processed/train.parquet')
test_path = Path('data/processed/test.parquet')

print('Train exists:', train_path.exists())
print('Test exists :', test_path.exists())

train = pd.read_parquet(train_path)
test = pd.read_parquet(test_path)

print('Train shape:', train.shape)
print('Test shape :', test.shape)

display(train.head())
```

```python
# Detect target column

def detect_target(df):
    for c in ['default','target','y','label','class','status']:
        if c in df.columns:
            return c
    return None

target = detect_target(train)
print('Detected target:', target)

if target is not None:
    print('Value counts (train):')
    print(train[target].value_counts(dropna=False))
    sns.countplot(x=target, data=train)
    plt.title('Target distribution (train)')
    plt.show()

# Missingness and dtypes
print('\nColumn dtypes:')
print(train.dtypes.value_counts())

miss = train.isna().mean().sort_values(ascending=False)
print('\nTop missingness:')
print(miss[miss>0].head(10))
```

```python
# Numeric summary for top numeric columns
num = train.select_dtypes(include=['number']).describe().T
display(num.head(20))

# Pairplot for small number of numeric features (sample)
nums = train.select_dtypes(include=['number']).columns.tolist()
sample_cols = nums[:6]  # limit to first 6 numeric features to keep plot small
if len(sample_cols) > 1:
    sns.pairplot(train[sample_cols].sample(min(500, len(train))), corner=True)
    plt.suptitle('Pairplot (sample)')
    plt.show()
```
